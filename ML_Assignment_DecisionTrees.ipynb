```python
from sklearn.preprocessing import LabelEncoder                 #importing essential libraries
```

```python
df1 = df[['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']]

for col in df1.columns:
  enc = LabelEncoder()                                        #using labelEncoder() function to convert string data to numerical data values
  enc.fit(df1[str(col)])
  df1[str(col)] = enc.transform(df1[str(col)])

```

```python
df1.groupby(by=df['class']).mean().plot.bar(alpha=0.5,figsize=(25,10))          #plotting a bar graph using grouby function on class attribute  
```

```python
X = df1                                        #splitting the data into features (X) and target (Y) to train further
Y = df['class']
```

```python
X.shape               #checking its total number of rows & columns to see if its split properly
```

```python
Y.shape                #checking its total number of rows & columns to see if its split properly
```

```python
from sklearn.model_selection import train_test_split          #importing essential libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42)             #splitting the data into train and test portions 
```

```python
model = DecisionTreeClassifier()                   #building decision tree model 
model = model.fit(X_train, y_train)                 #applying the decision tree model to the training data

predictions = model.predict(X_test)                  #predicting using test_data on the basis of model performance on training data

print("Score:", round(model.score(X_test, y_test),2)*100 ,'%')      #checking the score by comparing the feature test data to target test data and compute accuracy

```

```python
from IPython.display import Image
from six import StringIO  
import pydot                                                           #importing essential libraries

dot_data = StringIO()  
tree.export_graphviz(model, out_file=dot_data, feature_names=features)                        #passing the feature names of the df 
(graph,) = pydot.graph_from_dot_data(dot_data.getvalue())

Image(graph.create_png())                                               #plotting the image of the tree 
```
